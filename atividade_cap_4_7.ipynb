```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense #keras.layers.core tá dando erro
drive.mount('/content/drive')
TARGET_VARIABLE = "user_action" #variavel alvo
TRAIN_TEST_SPLIT = 0.5 # conjunto de treinamento
HIDDEN_LAYER_SIZE = 30 # tamanho da camada oculta
raw_data = pd.read_csv("/content/drive/MyDrive/Física - Bacharel Lic./deep learning & colisões de partículas/DL - Sandro/data.csv")
```

```python
mask = np.random.rand(len(raw_data)) < TRAIN_TEST_SPLIT # amostragem aleatoria dos dados para criar a divisão
tr_dataset = raw_data[mask] # treino
te_dataset = raw_data[~mask] # teste
```

```python
tr_data = np.array(tr_dataset.drop(columns=[TARGET_VARIABLE])) ## axis = 1 nao estava indo
tr_labels = np.array(tr_dataset[[TARGET_VARIABLE]])
te_data = np.array(te_dataset.drop(columns=[TARGET_VARIABLE]))
te_labels = np.array(te_dataset[[TARGET_VARIABLE]])
```

```python
## Especificação do modelo de rede neural no Keras
ffnn = Sequential() # inicializa um modelo sequencial na variavel ffnn
ffnn.add(Dense(HIDDEN_LAYER_SIZE, input_shape = (3, ), activation="sigmoid")) ## especifica tanto a camada de entrada quanto o tamanho da camada oculta
ffnn.add(Dense(1, activation="sigmoid")) # pega o tamanho da camada oculta da camada anterior e cria uma camada de saída com um neurônio
ffnn.compile(loss="mean_squared_error", optimizer="sgd", metrics=['accuracy']) # especifica a função de erro MSE, o otimizador e as metricas
ffnn.fit(tr_data, tr_labels, epochs = 150, batch_size=2, verbose=1) # treina a rede neural
```

```python
# Analisando os resultados no conjunto de teste
metrics = ffnn.evaluate(te_data, te_labels, verbose = 1) # avalia o modelo
print("%s: %.2f" % (ffnn.metrics_names[1], metrics[1]*100))
```
