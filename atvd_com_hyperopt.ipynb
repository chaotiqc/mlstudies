```python
def DNN(params):

    #reading parameters
    dim_input = params['dim_input'] # dimensao de entrada
    num_classes = params['num_classes'] # numero total de classes para a classificacao
    num_layers = params['num_layers'] #numero de camadas internas
    neurons = params['neurons'] #funcoes de ativacao
    activ = params['activ'] # funcao de ativacao para a camada oculta
    dp = params['dp'] # taxa de dropout (prob de desligar um neuronio)


    #camada de entrada
    input_data = Input(shape=(dim_input,), name='input_data') #recebe os dados, shape = numero de colunas ## shape=(dim_input,) especifica que o modelo espera vetores de entrada
    #camada interna
    x = Dense(int(neurons), activation=activ, name='layer_1')(input_data) #recebe os inputs e calculando a funcao de ativacao
    for k in range(1,num_layers-1):
        nome = 'layer_'+str(k+1)
        x = Dense(int(neurons), activation=activ, name=nome)(x) #recebe os inputs e calculando a funcao de ativacao
        x = BatchNormalization()(x) # melhorar performance (estabiliza e acelera o treinamento)
        x = Dropout(dp)(x) # melhorar, performance, desliga alguns que nao estao muito treinados (impede aue os neuronios dependa um do outro)

    #ultima camada
    if num_classes==2:
        xo = Dense(num_classes-1, activation='sigmoid')(x) # se tiver apenas 2 classes sigmoid, nosso caso, 1 neuron na ultima camada
    else:
        xo = Dense(num_classes, activation='softmax')(x) # mais de duas use softmax

    model = Model(inputs=input_data, outputs=xo)

    return model
```

```python
import tensorflow as tf

#Building the Classifier #
def classifier_DNN_builder(feats,labels,params,opt,lr,epochs,batch_size):

    X_train, _, X_val = feats
    y_train, _, y_val = labels
    sel_train = selector(y_train)
    sel_eval  = selector(y_val)

    # number of S,B MC events in the sample
    NS=len(y_train[sel_train[1]])
    NB=len(y_train[sel_train[0]])


    # fitting the model
    eval_set = [X_val, y_val]
    ###############################################
    checkpoint_filepath = '/content/drive/MyDrive/Física - Bacharel Lic./deep learning & colisões de partículas/best_weights.weights.h5'
    #salvar o melhor modelo, melhor modelo nem sempre é o modelo da ultima epoca
    model_checkpoint_callback = ModelCheckpoint(
                    filepath=checkpoint_filepath,
                    save_weights_only=True,
                    monitor='val_loss',
                    mode='min',
                    save_best_only=True)

    # instantiate the model
    model = DNN(params)


    opt = opt
    #opt = tfa.optimizers.AdamW(learning_rate=lr, weight_decay=wd)
    # para melhorar o learning rate, foca na regiao do minimo
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                              patience=25, min_lr=1e-6) #factor = 0.5 diminui na metade o LR apos 25 epocas



    ########## Early stopping
    stop = EarlyStopping(monitor='val_loss', mode='min', patience=25)

    ######################################
    # Compilation
    model.compile(loss='binary_crossentropy',optimizer=opt, metrics=['accuracy'])
    model.summary()

    # fitting the model
    history=model.fit(x=X_train,y=y_train,
                    epochs=epochs,
                    batch_size=batch_size,
                    callbacks=[model_checkpoint_callback, reduce_lr, stop], #,TensorBoard(log_dir='/tmp/autoencoder')],
                    #validation_split = 0.25, #faz o que a linha de debaixo faz
                    validation_data=(eval_set),
                    verbose = 1)

    #model.save_weights('/content/drive/MyDrive/Física - Bacharel Lic./deep learning & colisões de partículas/class_weights.h5')
    #tava dando erro a linha acima

    return model, history

```

```python
# AMS metrics (Casos extremos, eu acho)
def ams(s,b,sys):
    breg=0.
    if b==0. and sys!=0.:
        aux=np.sqrt(2*s*(1.+1./sys))
    else:
        b=b+breg
        if sys==0.:
            aux=np.sqrt(2.0)*np.sqrt(-s+(s+b)*np.log(1.+s/b))
        else:
            aux=np.sqrt(2.0)*np.sqrt((s+b)*np.log((1.+1./(b*sys**2))*(s+b)/(s+b+1./sys**2))+(1./sys)**2*np.log((b+1./sys**2)/(s+b+1./sys**2)))
    return aux
```

```python
def amscut(lumi,sys,wS,wB,scores_S,scores_B):

    vres = list() # Vai guardar o resultado geral
    veffS =list() # Vai guardar a eficiência de sinal
    veffB =list() # Vai guardar a rejeição de background.

# Variaveis auxiliares :
    sm=0.
    bm=0.
    aux=0.

    for k in range(0,1000): # loop de 0  a 999.
        cut=0.+0.001*k # vai deixar o corte entre 0 e 1.
        # filtro. se a condição for maior que o corte, calcular a Luminosidade X Peso
        fS =[lumi*wS for i in range(len(scores_S)) if abs(scores_S[i]-0.0)>cut]
        fB =[lumi*wB for i in range(len(scores_B)) if abs(scores_B[i]-0.0)>cut]

        # Metricas de performance
        effS =float(len(fS))/len(scores_S) # Divide o número de eventos que passaram (len(fS)) pelo total original
        effB =float(len(fB))/len(scores_B) # Divide o número de eventos que passaram (len(fB)) pelo total original
        rejB =1.-effB # O inverso da eficiência do back

        # ​Adiciona os valores calculados neste passo às listas acumuladoras usando numpy
        veffS =np.append(veffS,effS)
        veffB =np.append(veffB,rejB)

        s =np.sum(fS) # soma de todos os pesos da lista fS
        b =np.sum(fB) # soma de todos os pesos da lista fB

        #b=b1+b2+b3+b4
        sig=ams(s,b,sys) # ​Chama a função ams passando os totais s e b e a incerteza sys


        vaux=np.array([cut,s,b,sig]) #empacota os 4 dados importantes desse passo (cut, s, b, sig) num array pequeno vaux
        vres=np.append(vres,vaux) # uma lista mais geral
    return vres.reshape(1000,4), veffS, veffB # cria um array unidunidimensional gigante
```

```python
def DNN(params):

    #reading parameters
    dim_input = X_val.shape[1]
    num_classes = params['num_classes']
    num_layers = params['num_layers'] #numero de camadas internas
    neurons = params['neurons'] #funcoes de ativacao
    activ = params['activ']
    dp = params['dp']

    # building the neural net

    #camada de entrada
    input_data = Input(shape=(dim_input,), name='input_data') #recebe os dados, shape = numero de colunas
    #camada interna
    x = Dense(int(neurons), activation=activ, name='layer_1')(input_data) #recebe os inputs e calculando a funcao de ativacao
    for k in range(1,int(num_layers)-1):
        nome = 'layer_'+str(k+1)
        x = Dense(int(neurons), activation=activ, name=nome)(x) #recebe os inputs e calculando a funcao de ativacao
        x = BatchNormalization()(x) # melhorar performance
        x = Dropout(dp)(x) # melhorar, performance, desliga alguns que nao estao muito treinados
    #ultima camada
    if num_classes==2:
        xo = Dense(num_classes, activation='sigmoid')(x) # se tiver apenas 2 classes sigmoid, nosso caso
    else:
        xo = Dense(num_classes, activation='softmax')(x) # mais de duas use softmax

    model = Model(inputs=input_data, outputs=xo)

    return model
```

```python
def DNN_tuning(space):

    # instantiate the model
    model = DNN(space)
    opt = Adam(learning_rate=space['learning_rate']) # Configura o otimizador Adam. Ele pega a learning_rate específica que o Hyperopt escolheu para esta rodada.
    reduce_lr = ReduceLROnPlateau(monitor='lr', factor=0.5, # Cria um callback para reduzir a taxa de aprendizado se o treino estagnar
                              patience=20, min_lr=1e-9)

    ########## Early stopping
    stop = EarlyStopping(monitor='val_loss', mode='min', patience=25) # Se a perda na validação (val_loss) não cair (melhorar) por 25 épocas seguidas (patience=25), o treino para

    #####################################
    model.compile(loss='binary_crossentropy',optimizer=opt, metrics=['accuracy'])

    ntr = int(0.5*len(y_val)) #uso para treinar metade do conjunto de validacao # Calcula o índice que representa 50% do tamanho do seu conjunto de validação (y_val)


    history = model.fit(x=X_val[:ntr],y=tf.one_hot(y_val[:ntr], depth=2), # Usa apenas a primeira metade dos seus dados de validação como entrada para treinar este teste . ## ​y=tf.one_hot(...): Converte seus targets (0 e 1) para o formato One-Hot ([1,0] e [0,1]). Isso indica que sua rede DNN provavelmente tem 2 neurônios na saída (Softmax).
                epochs=50,
                batch_size=int(space['batch_size']), # Converte o número decimal do Hyperopt (ex: 32.0) para inteiro (32), pois o fit exige inteiros.
                callbacks=[reduce_lr, stop],
                validation_split = 0.25, # Separa automaticamente 25% desses dados (da metade usada) para calcular a val_loss durante o treino.
                verbose = 0)

    val_loss = np.amin(history.history['val_loss']) #  Olha para o histórico de todo o treinamento e pega o menor valor que a val_loss atingiu em qualquer época. Esse é o "placar" desse modelo.
    print('Best validation loss of epoch:', val_loss)

    return {'loss': val_loss, 'status': STATUS_OK, 'model': model}
```

```python
trials = Trials() # Inicializa um objeto que vai servir como um banco de dados em memória.
best = fmin(fn=DNN_tuning,  # ​fmin: Significa "Function Minimization" (Minimização de Função). O objetivo do Hyperopt é encontrar o valor mínimo da loss que a sua função DNN_tuning retorna.
            space=space,
            algo =tpe.suggest, # TPE significa Tree-structured Parzen Estimator é um metodo Bayesiano.
            max_evals=200,
            trials=trials)

print(best)
```
